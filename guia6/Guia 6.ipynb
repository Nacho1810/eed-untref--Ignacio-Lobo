### Ejercicio 6

Sea $G$ un grafo no dirigido, utilizar el recorrido DFS para determinar si $G$ es bipartito.

```python
from edd.grafo import Grafo, Vertice
from edd.cola import Cola
def es_bipartito_con_dfs(G: Grafo,s:Vertice):  # agregar argumentos necesarios.
    
    s = G._vertices[str(s)]
    novisit = {v:True for v in G.vertices}
    color = {}
    color[s]= True
    c = Cola()
    c.encolar(s)
    while not c.esta_vacia():
        v = c.desencolar()
        for w in v.adyacentes:
            if novisit[w]:
                novisit[w]=False
                color[w] = not color[v]
                c.encolar(w)
            elif color[w] == color[v]:
                return False
            
    return True

G = Grafo()
G.agregar_arista("B", "D")
G.agregar_arista("B", "D")
G.agregar_arista("C", "A")
G.agregar_arista("A", "B")
G.draw()
print(es_bipartito_con_dfs(G,"A"))
G.agregar_arista("C","E")
G.agregar_arista("D","E")

G.draw()
print(es_bipartito_con_dfs(G,"A"))
```
