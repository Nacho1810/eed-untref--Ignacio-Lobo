### Ejercicio 4

Escribir un método que imprima por pantalla el camino mínimo y el costo para llegar a cada vértice luego de ejecutar los algoritmos de camino mínimo.

```python
from edd.grafo import DiGrafo

# made in nacho
def imprimir_caminos(self,s) -> None:

    dic_distancias,dic_previos = self.dijkstra(s)
    impresion=[]
    for vertice,distancia in dic_distancias.items():
        pila = []
        #vertice != None
        while vertice:
            pila.append(vertice.id)
            vertice= dic_previos[vertice]
        impresion.append("-".join(pila[::-1]) + " " + str(distancia))
        print("-".join(pila[::-1]),distancia)
    return "\n".join(impresion)

DiGrafo.imprimir_caminos = imprimir_caminos

# Salida de ejemplo:
# A-C-D 15
# A-B 7
# A-B-E 10
# A-D 3
# A 0

G = DiGrafo()
G.agregar_arista("A", "B", 7)
G.agregar_arista("A", "C", 8)
G.agregar_arista("C", "D", 7)
G.agregar_arista("B", "E", 3)
G.agregar_arista("E", "C", 9)

# d,p = G.dijkstra("A")

G.imprimir_caminos("A")
G.draw()
```

### Ejercicio 11

Modificar el algoritmo de Dikstra para que en el caso que haya más de un camino mínimo hasta un vértice registre todos los caminos mínimos

```python
from edd.grafo import DiGrafo, ColaDePrioridad

def dijkstra_caminos_minimos(self,s,func):

    s = self._vertices[str(s)]

    pq = ColaDePrioridad()

    distancias = {v:inf for v in self.vertices}
    previos = {v:None for v in self.vertices}
    visitados = {v:False for v in self.vertices}
    
    distancias[s] = 0
    pq.encolar(s,0)

    while not pq.esta_vacia():
        v,_ = pq.desencolar_minimo()
        func(v)
        visitados[v] = True

        for _, w, peso_v_w in v.aristas:

            if not visitados[w] and distancias[w] >= distancias[v] + peso_v_w:
                if distancias[w] > distancias[v]+ peso_v_w:
                    distancias[w] = distancias[v] + peso_v_w
                    previos[w] = [v]
                    pq.encolar(w,distancias[w])
                else:
                    previos[w].append(v)


    return distancias, previos

DiGrafo.dijkstra_caminos_minimos = dijkstra_caminos_minimos

G = DiGrafo()
G.agregar_arista("A", "B", 5)
G.agregar_arista("B", "C", 1)
G.agregar_arista("A", "C", 6)
G.agregar_arista("B", "D", 3)
G.agregar_arista("B", "E", 4)
G.agregar_arista("C", "E", 3)
G.agregar_arista("A", "D", 9)
G.agregar_arista("A", "E", 9)


print(G.dijkstra_caminos_minimos("A",print))
print(G.dijkstra_caminos_minimos("B",print))
G.draw()
```
